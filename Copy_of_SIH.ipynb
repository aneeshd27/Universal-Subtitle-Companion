```python
import moviepy.editor as mpe

# Specify the path to the MP4 file
video_path = "video.mp4"

# Create a VideoFileClip object from the MP4 file
video = mpe.VideoFileClip(video_path)

# Extract the audio from the MP4 file
audio = video.audio

# Specify the output audio file path with the .wav extension
audio_path = "audio.wav"

# Write the audio to the specified file (which will be in WAV format)
audio.write_audiofile(audio_path)
```

```python
# Transcribe the audio to obtain segments with words and their timestamps
from faster_whisper import WhisperModel

# Load the Whisper ASR model
model_size = "small"
model = WhisperModel(model_size)

# Transcribe the audio and get word timestamps
segments, _ = model.transcribe("audio.wav", word_timestamps=True)

# Create the SRT file content from the transcribed segments
def generate_srt_content(segments):
    srt_content = ""
    counter = 1

    for segment in segments:
        start_time = segment.start
        end_time = segment.end
        text = " ".join(word.word for word in segment.words)

        # Convert start and end times to HH:MM:SS,ms format
        start_time_str = "{:02d}:{:02d}:{:06.3f}".format(int(start_time // 3600), int((start_time % 3600) // 60), start_time % 60)
        end_time_str = "{:02d}:{:02d}:{:06.3f}".format(int(end_time // 3600), int((end_time % 3600) // 60), end_time % 60)

        # Create the subtitle block
        srt_content += f"{counter}\n{start_time_str} --> {end_time_str}\n{text}\n\n"
        counter += 1

    return srt_content

# Generate the SRT file content
srt_content = generate_srt_content(segments)

# Specify the output subtitle file path with the .srt extension
subtitle_output_path = "english_subtitles.srt"

# Write the SRT content to the specified file
with open(subtitle_output_path, 'w') as f:
    f.write(srt_content)

print("Subtitle .srt file generated successfully:", subtitle_output_path)
```

```python
import subprocess

def burn_subtitles(video_path, subtitle_path, output_video_path):
  """Burns subtitles on a video file.

  Args:
    video_path: The path to the video file.
    subtitle_path: The path to the subtitle file.
    output_video_path: The path to the output video file.
  """

  # Create a FFmpeg command.
  ffmpeg_cmd = ["ffmpeg", "-i", video_path, "-vf", f"subtitles={subtitle_path}", output_video_path]

  # Run the FFmpeg command.
  subprocess.run(ffmpeg_cmd)

# Get the paths to the video and subtitle files.
video_path = "video.mp4"
subtitle_path = "english_subtitles.srt"

# Get the output video file path.
output_video_path = "output_eng.mp4"

# Burn the subtitles on the video file.
burn_subtitles(video_path, subtitle_path, output_video_path)
```

```python
import pysrt
from googletrans import Translator

# Load English subtitles
subtitle_path = 'english_subtitles.srt'
subtitles = pysrt.open(subtitle_path, encoding='utf-8')

# Initialize the translator
translator = Translator()

# Translate each subtitle and create Marathi subtitles
marathi_subtitles = []

for subtitle in subtitles:
    # Translate the subtitle text
    translation = translator.translate(subtitle.text, src='en', dest='mr')
    translated_text = translation.text

    # Retain the original timing
    marathi_subtitle = pysrt.SubRipItem(
        start=subtitle.start,
        end=subtitle.end,
        text=translated_text
    )
    marathi_subtitles.append(marathi_subtitle)

# Save Marathi subtitles to a new file
marathi_subtitle_path = 'marathi_subtitles.srt'
with open(marathi_subtitle_path, 'w', encoding='utf-8') as marathi_file:
    marathi_file.write('\n'.join([str(sub) for sub in marathi_subtitles]))

print('Marathi subtitles saved successfully.')
```
